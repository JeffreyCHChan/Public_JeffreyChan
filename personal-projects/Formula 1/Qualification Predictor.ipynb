```python
fastf1.Cache.enable_cache("./cache")

event_name = "Austrian"
session = fastf1.get_session(2023, event_name, 'FP1')
session.load()
laps = session.laps

session_name = session.name
```

```python
col = session.laps

col.drop(['DriverNumber','PitOutTime', 'PitInTime','LapStartTime',
       'LapStartDate', "Position", "Sector1SessionTime", "Sector2SessionTime", "Sector3SessionTime"], axis=1, inplace=True)
col["DeletedReason"].fillna("NotDeleted", inplace=True)
```

```python

# col.drop(col[col["LapTime"] == pd._libs.tslibs.nattype.NaTType].index, inplace=True)
col.dropna(inplace=True)
col
```

```python
# laps = session.laps

# print(type(col.Time[0]))
# print(type(col.Time))
pd.to_timedelta(col["Time"])

for column in ['LapTime','Sector1Time',
    'Sector2Time', 'Sector3Time']:

    col[column] = col[column].dt.total_seconds()*1000

    # Remove the intermediate column if desired
    # col = col.drop(columns=["Temp"])

# TODO pandas is having difficulty converting Timedelta data type

col
```

```python
col["temp"] = col["Time"].dt.total_seconds().astype(int)

# Convert the `timedelta_column_seconds` to `datetime64[ns]` by adding it to a reference date
col["Time"] = pd.to_datetime(session.date) + pd.to_timedelta(col["temp"], unit="s")

# Remove the intermediate column if desired
col = col.drop(columns=["temp"])
```
